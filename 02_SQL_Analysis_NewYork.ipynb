# 🏠 New York Airbnb SQL Analysis

## 📖 Objective

This notebook uses SQL to answer business questions related to Airbnb listings in New York. The analysis focuses on pricing, occupancy, revenue, customer ratings, host performance, and neighborhood trends to generate actionable business insights.

# 📊 Section 1: Market Overview

This section provides an overview of the New York Airbnb market, including the total number of listings, average pricing, estimated revenue, and the distribution of different accommodation types.

## 📌 Query 1: Total Airbnb Listings
## 🎯 Business Question

How many Airbnb listings are available in New York?

```python
query = """
SELECT COUNT(*) AS total_listings
FROM airbnb_listings;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

The dataset contains 30,259 Airbnb listings, providing a large sample for analyzing pricing, revenue, occupancy, and customer behavior across New York.

## 📌 Query 2: Average Nightly Price

### 🎯Business Question
What is the average nightly price of an Airbnb listing?

```python
query = """
SELECT ROUND(AVG(price),2) AS average_price
FROM airbnb_listings;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

The average nightly price serves as a benchmark for understanding the overall pricing level of Airbnb accommodations in New York and provides a baseline for comparing room types and neighborhoods.

## 📌 Query 3: Average Estimated Annual Revenue

### 🎯 Business Question

What is the average estimated annual revenue generated by Airbnb listings?

```python
query = """
SELECT ROUND(AVG(estimated_revenue_l365d),2) AS average_revenue
FROM airbnb_listings;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

The average estimated annual revenue highlights the overall earning potential of Airbnb listings and helps evaluate the profitability of the New York Airbnb market.

## 📌 Query 4: Distribution of Room Types

### 🎯 Business Question

Which room types are most commonly listed on Airbnb in New York?

```python
query = """
SELECT room_type,
       COUNT(*) AS total_listings
FROM airbnb_listings
GROUP BY room_type
ORDER BY total_listings DESC;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Entire homes/apartments represent the largest share of Airbnb listings, followed by private rooms, while hotel rooms and shared rooms account for a much smaller portion of the market.

## 📌 Query 5: Average Price by Room Type

### 🎯 Business Question

Which room type has the highest average nightly price?

```python
query = """
SELECT room_type,
       ROUND(AVG(price),2) AS average_price
FROM airbnb_listings
GROUP BY room_type
ORDER BY average_price DESC;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Hotel rooms have the highest average nightly price, while shared rooms are the most affordable accommodation option. This reflects the differences in amenities, privacy, and guest experience offered by each room type.

## 📌 Query 6: Neighborhood Coverage

### 🎯 Business Question

How many unique neighborhoods are represented in the Airbnb dataset?

```python
query = """
SELECT COUNT(DISTINCT neighbourhood_cleansed) AS total_neighbourhoods
FROM airbnb_listings;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

The dataset covers a wide range of neighborhoods across New York, allowing for comprehensive location-based pricing and revenue analysis.

## 📌 Query 7: Property Type Diversity

### 🎯 Business Question

How many different property types are listed on Airbnb?

```python
query = """
SELECT COUNT(DISTINCT property_type) AS total_property_types
FROM airbnb_listings;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

The variety of property types demonstrates the diversity of accommodations available on Airbnb, ranging from private rooms to luxury homes and unique stays.

# 💰 Section 2: Pricing Analysis

This section analyzes Airbnb pricing patterns across different room types, neighborhoods, property types, and host categories. The goal is to identify factors that influence listing prices and understand market trends.

## 📌 Query 8: Top 10 Most Expensive Neighborhoods

### 🎯 Business Question

Which neighborhoods have the highest average Airbnb prices?

```python
query = """
SELECT neighbourhood_cleansed,
       ROUND(AVG(price),2) AS average_price
FROM airbnb_listings
GROUP BY neighbourhood_cleansed
ORDER BY average_price DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Luxury neighborhoods command significantly higher nightly prices than other areas. Understanding these premium markets helps identify high-value investment opportunities and pricing strategies.

## 📌 Query 9: Average Price by Property Type

### 🎯 Business Question

Which property types have the highest average nightly price?

```python
query = """
SELECT property_type,
       ROUND(AVG(price),2) AS average_price
FROM airbnb_listings
GROUP BY property_type
ORDER BY average_price DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Luxury and unique accommodation types generally command higher prices, while standard rental units and private rooms remain more affordable for travelers.

## 📌 Query 10: Average Price by Bedrooms

### 🎯 Business Question

How does the average nightly price vary with the number of bedrooms?

```python
query = """
SELECT bedrooms,
       ROUND(AVG(price),2) AS average_price
FROM airbnb_listings
GROUP BY bedrooms
ORDER BY bedrooms;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Listings with more bedrooms generally command higher nightly prices, reflecting their ability to accommodate larger groups and offer additional space.

## 📌 Query 11: Superhost Pricing Comparison

### 🎯 Business Question

Do Superhosts charge higher prices than non-Superhosts?

```python
query = """
SELECT host_is_superhost,
       ROUND(AVG(price),2) AS average_price
FROM airbnb_listings
GROUP BY host_is_superhost
ORDER BY average_price DESC;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Comparing average prices by Superhost status helps determine whether host reputation influences pricing strategies and guest willingness to pay.

## 📌 Query 12: Most Expensive Airbnb Listings

### 🎯 Business Question

Which Airbnb listings have the highest nightly prices?

```python
query = """
SELECT id,
       neighbourhood_cleansed,
       property_type,
       room_type,
       price
FROM airbnb_listings
ORDER BY price DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

Identifying the most expensive listings highlights the luxury segment of the Airbnb market and provides insight into premium pricing across different neighborhoods and property types.

# ⭐ Section 4: Customer & Host Analysis

This section analyzes guest reviews, host performance, and Superhost status to understand customer satisfaction and the impact of host reputation on Airbnb listings.

## 📌 Query 18: Average Rating by Room Type

### 🎯 Business Question

Which room type receives the highest average guest rating?

```python
query = """
SELECT room_type,
       ROUND(AVG(review_scores_rating),2) AS average_rating
FROM airbnb_listings
GROUP BY room_type
ORDER BY average_rating DESC;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Comparing guest ratings across room types helps identify which accommodation categories consistently provide better guest experiences.

## 📌 Query 19: Superhost Performance

### 🎯 Business Question

Do Superhosts receive higher guest ratings than non-Superhosts?

```python
query = """
SELECT host_is_superhost,
       ROUND(AVG(review_scores_rating),2) AS average_rating
FROM airbnb_listings
GROUP BY host_is_superhost
ORDER BY average_rating DESC;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Superhost ratings provide insight into whether experienced and highly rated hosts consistently deliver better guest experiences.

## 📌 Query 20: Most Reviewed Neighborhoods

### 🎯 Business Question

Which neighborhoods have the highest average number of guest reviews?

```python
query = """
SELECT neighbourhood_cleansed,
       ROUND(AVG(number_of_reviews),2) AS average_reviews
FROM airbnb_listings
GROUP BY neighbourhood_cleansed
ORDER BY average_reviews DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Neighborhoods with higher review counts often indicate stronger guest demand and more established Airbnb activity.

## 📌 Query 21: Reviews by Room Type

### 🎯 Business Question

Which room types receive the highest average number of guest reviews?

## 📌 Query 22: Highest Rated Neighborhoods

### 🎯 Business Question

Which neighborhoods have the highest average guest ratings?

```python
query = """
SELECT neighbourhood_cleansed,
       ROUND(AVG(review_scores_rating),2) AS average_rating
FROM airbnb_listings
GROUP BY neighbourhood_cleansed
HAVING COUNT(review_scores_rating) >= 10
ORDER BY average_rating DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### 📈 Business Insight

Filtering neighborhoods with at least 10 ratings helps ensure that the results are based on a meaningful number of guest reviews rather than a few exceptional cases.
